## Word2Vec 이란?

image.png

Word2Vec은 이 임베딩을 만들고 학습하는 가장 오래된 방법 중 하나이다. Word2Vec은 단일한 알고리즘이 아니라 대규모 데이터 세트에서 단어 임베딩을 학습하는 데 사용할 수 있는 모델 아키텍처 및 최적화 제품군이다. Word2Vec을 통해 학습된 임베딩은 텍스트 분류, 질문 답변과 같은 다양한 다운스트림 자연어 처리 작업에서 성공적인 것으로 입증되었다. 이 논문은 단어 표현을 학습하기 위한 두 가지 방법을 제안했다.

image.png

**Continuous Bag-of-Words Model(CBOW)** 주변 문맥 단어를 기반으로 중간 단어를 예측합니다. 문맥은 현재(중간) 단어 앞과 뒤에 있는 몇 개의 단어로 구성됩니다. 이 구조는 문맥에서 단어의 순서가 중요하지 않다.

image.png

**Skip-gram Model**  같은 문장에서 중심 단어 앞뒤로 특정 범위 내의 단어를 예측합니다.


image.png


**CBOW 모델과 Skip-gram 모델의 차이** CBOW는 skip-gram보다 학습 시간이 상대적으로 적게 걸리며, 자주 등장하는 단어에 대해 좋은 예측력을 보인다. 
Skip-gram은 자주 등장하지 않는 단어들에 대해서 CBOW보다 좋은 단어 예측력을 보인다

## 3. 전처리 <a class="anchor" id="3"></a>
NLTK 라이브러리에서 word_tokenize를 사용하여 모든 텍스트의 소문자를 제거하고 텍스트를 토큰화(개별 토큰/단어로 분할)합니다.



```python
import nltk
nltk.download('stopwords')
nltk.download('punkt')
```

```python
def preprocess_corpus(texts):
    #nltk에서 다운로드 받은 stopwords를 삭제한다. 
    mystopwords = set(stopwords.words("english"))
    def remove_stops_digits(tokens):
        return [token.lower() for token in tokens if token not in mystopwords and not token.isdigit()
               and token not in punctuation]
    #위에서 처리한 내용을 토큰화 
    return [remove_stops_digits(word_tokenize(text)) for text in tqdm(texts)]

#train test data에 모두 적용
train_texts_processed = preprocess_corpus(train_texts)
test_texts_processed = preprocess_corpus(test_texts)
print("Example Training Prepocessed Text\n\n", train_texts_processed[0])
```

## 5. 사전 훈련된 모델을 사용하여 텍스트 입력을 임베딩으로 변환 <a class="anchor" id="5"></a>

여기서 우리는 이전의 입력 토큰화된 텍스트를 가져와서 사전 훈련된 임베딩 벡터로부터 텍스트의 각 단어에 대한 임베딩을 얻는다. 이는 출력 변수와 함께 훈련하는 데 사용할 수 있는 문장당 임베딩 형태의 최종 입력 데이터 세트를 제공한다.

```python
#각 문장이 단어 목록인 목록 형태로 입력 텍스트 데이터 세트를 가져가는 함수
def embedding_feats(list_of_lists, DIMENSION, w2v_model):
    zeros_vector = np.zeros(DIMENSION)
    feats = []
    missing = set()
    missing_sentences = set()
    #문장마다 이동
    for tokens in tqdm(list_of_lists):
        # 처음에는 문장의 임베딩 벡터로 0을 지정한다.
        feat_for_this = zeros_vector
        #이 문장의 임베딩에 포함된 단어 수를 카운트한다.
        count_for_this = 0
        for token in tokens:
            #단어가 임베딩 벡터에 있는지 확인
            if token in w2v_model:
                #문장의 벡터에 단어의 벡터를 추가합니다.
                feat_for_this += w2v_model[token]
                count_for_this +=1
            #Else assign the missing word to missing set just to have a look at it
            else:
                missing.add(token)
        #없어진 단어를 보기 위해 누락된 집합에 지정한다.
        if count_for_this == 0:
            #해당 문장에 대해 모든 영점 벡터를 지정한다.
            feats.append(feat_for_this)
            #누락된 문장을 missing_sentences에 추가한다.
            missing_sentences.add(' '.join(tokens))
        #문장의 임베딩을 얻기 위해 각 단어에 대한 임베딩 값의 평균을 구하라.
        else:
            feats.append(feat_for_this/count_for_this)
    return feats, missing, missing_sentences
```

```python
#Embeddings for the train dataset
train_vectors, missing, missing_sentences = embedding_feats(train_texts_processed, 300, w2v_google_news)
```

## 6. Multi-Output Classifier 학습과 검증 <a class="anchor" id="6"></a>
우리는 각 문장을 oxic or not, severe_toxic or not, obscene or not, threat or not, insult or not and identity_hate or not 우리는 6개의 출력 변수에 대해 문장을 분류해야 한다. 이것은 대상 변수가 2개 이상의 옵션을 갖는 다중 클래스 분류와 다르다. 예: 문장은 양수, 음수 및 중립일 수 있음)

마찬가지로, 앞에서 언급한 바와 같이 래퍼인 sklearn의 MultiOutputClassifier를 사용할 예정입니다.

1. 교육 데이터 세트에 대한 임베딩 벡터 가져오기
2. 임베딩 벡터 및 출력 변수를 열차 및 검증 세트로 분할
3. 임베딩 벡터 및 출력 교육에 로지스틱 회귀 모형 적합variables
4. 검증 임베딩 벡터에 대한 예측
5. ROC-AUC의 관점에서 성능 측정

image.png

roc_auc_measure에 의해 사용되기 때문에 0.5의 임계값을 기반으로 예측 값 대신 확률 점수를 제공하는 예측 대신 모델의 pedict_proba 함수를 사용할 것입니다.

```python
#대상변수에 걸쳐 주어진 실제 이항값과 모델에 의한 확률점수를 계산하기 위한 함수
def accuracy(y_test, y_pred):
    aucs = []
    #Calculate the ROC-AUC 
    for col in range(y_test.shape[1]):
        aucs.append(roc_auc_score(y_test[:,col],y_pred[:,col]))
    return aucs
```

```python
def train_model(DIMENSION, model):
    train_vectors, missing, missing_sentences = embedding_feats(train_texts_processed, DIMENSION, model)
    
    #훈련 데이터의 임베딩 벡터를 출력 변수와 함께 훈련 및 검증 세트로 분할

    train_data, val_data, train_cats, val_cats = train_test_split(train_vectors, train_labels)
    
    #Logistic Regression Model
    lr = MultiOutputClassifier(LogisticRegression(class_weight='balanced', max_iter=10000)).fit(train_data, train_cats)


    y_vals = val_cats
    y_preds = np.transpose(np.array(lr.predict_proba(val_data))[:,:,1])
    #Calculate the Mean ROC_AUC 
    mean_auc = mean(accuracy(y_vals,y_preds))
    return mean_auc, lr
```

```python
mean_auc, lr = train_model(300, w2v_google_news)
print(mean_auc)
```
